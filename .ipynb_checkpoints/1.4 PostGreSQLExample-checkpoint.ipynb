```python
from sqlalchemy import Column, Integer, String, Boolean, Numeric, ForeignKey, Table
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import relationship

Base = declarative_base()

users_products_association = Table('users_products', Base.metadata,
    Column('user_id', Integer, ForeignKey('users.id')),
    Column('product_id', Integer, ForeignKey('products.id'))
)


class Product(Base):
    __tablename__ = 'products'
    id=Column(Integer, primary_key=True)
    title=Column('title', String(32))
    price=Column('price', Numeric)
    comments = relationship("Comment", back_populates="product")
    publisher_id = Column(Integer, ForeignKey('users.id'))
    publisher = relationship("User", foreign_keys=[publisher_id], back_populates="product_published") # , uselist=False , 
    owners = relationship("User", secondary=users_products_association)
    def __init__(self, title, publisher, price=10):
        self.title = title
        self.publisher = publisher
        self.price = price

class Comment(Base):
    __tablename__ = 'comments'
    id = Column(Integer, primary_key=True)
    content=Column('content', String(100))    
    product_id = Column(Integer, ForeignKey('products.id'))
    product = relationship("Product", back_populates="comments")
    def __init__(self, content):
        self.content = content

class User(Base):
    __tablename__ = 'users'
    id=Column(Integer, primary_key=True)
    name=Column('name', String(32))
    product_published = relationship("Product")
    products_bought = relationship("Product", secondary=users_products_association)
    def __init__(self, name):
        self.name = name
```
